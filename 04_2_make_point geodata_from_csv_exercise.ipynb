## 사용할 데이터 준비
공공데이터포털의 "전국문화축제표준데이터"를 활용합니다 

```python
url = 'https://www.data.go.kr/download/15013104/standard.do?dataType=json'
dict = json.loads(str(requests.get(url).text))
dict
```

```python
df = json_normalize(dict['records']) 
df.info()
df.head()
```

### 좌표 속성 정보가 포함되어 있는가요?   
### 있다면, 위도, 경도 항목을 이용하여 Point Geometry를 만들 수 있습니다. 

```python
# 좌표 항목을 가지고 지오메트리 데이터를 생성
geom = gpd.points_from_xy(df.경도, df.위도)
```

### Error의 원인이 무엇이고, 어떻게 해결하면 될까요?

```python
# 문자열 값에 들어있는 쉼표를 제거 - 쉼표에 대해서는 위의 정규표현식 특수문자 처리와 동일 효과
# df['경도'] = df['경도'].str.replace(',', '')
# df['위도'] = df['위도'].str.replace(',', '')
```

NUll 데이터가 문제라면, 아래 코드로 null인 데이터(행)을 제거하면 됩니다. 

```python
# 경도 좌표가 null 아닌 데이터만 추출
df2 = df.loc[ df.notnull()['경도'] , :]
# 아래 코드도 같은 기능을 하고, 새로운 객체를 만들 필요가 없습니다. 
#df.dropna(subset=['경도'], inplace=True)
```

### 불필요한 문자열이나, Null 데이터가 처리된 후에, 데이터타입을 실수형으로 변경합니다. 

```python
#데이터 타입 변환
df2['경도'] = pd.to_numeric(df2['경도']) 
df2['위도'] = pd.to_numeric(df2['위도']) 
```

```python
geom = gpd.******_from_xy(df2.경도, df2.위도)
```

## geom 도형 객체와 데이터프레임의 속성정보를 붙여서 GeoDataFrame을 만들 수 있습니다. 

```python
# 지오메트리를 붙인 공간데이터프레임 생성 
gdf = gpd.GeoDataFrame(df2, ********=geom, ***="epsg:4326")
gdf.info()
gdf.head()
```

#### 항목이 너무 많고, 한글이어서 데이터 가공/관리가 좀 불편합니다   
#### 먼저 필요한 항목만 추출해보겠습니다

```python
gdf2 = gdf[['축제명', '제공기관명', 'geometry']]
gdf2.head()
```

#### 항목명을 변경할 수 있습니다

```python
gdf2.columns = ['name', 'organization', 'geometry']
gdf2.head()
```
